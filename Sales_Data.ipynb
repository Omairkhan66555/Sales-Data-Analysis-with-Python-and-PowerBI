```python
import pandas as pd

df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
df.head(100)
```

```python
# Shape of the dataset (rows, columns)
print("Shape of dataset:", df.shape)

# Column names
print("\nColumns in dataset:\n", df.columns)

# Data types of each column
print("\nData types:\n", df.dtypes)

# Null/missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
df_cleaned = df.drop(columns=['ADDRESSLINE2', 'PHONE', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME', 'STATE'])
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.dropna()
df_cleaned.shape
```

```python
df_cleaned['Total Revenue'] = df_cleaned['QUANTITYORDERED'] * df_cleaned['PRICEEACH']
df_cleaned[['QUANTITYORDERED', 'PRICEEACH', 'Total Revenue']].head()
```

```python
top_products = df_cleaned.groupby('PRODUCTCODE')['Total Revenue'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
import matplotlib.pyplot as plt

top_products.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title('Top 10 Best Selling Products by Revenue')
plt.xlabel('Product Code')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig('top_10_products_bar.png')   # 🔸 Save image before plt.show()
plt.show()

from google.colab import files
files.download('top_10_products_bar.png')
```

```python
# Convert order date column to datetime
df_cleaned['ORDERDATE'] = pd.to_datetime(df_cleaned['ORDERDATE'])

# Create Year-Month column
df_cleaned['YearMonth'] = df_cleaned['ORDERDATE'].dt.to_period('M')

# Group by YearMonth and sum Total Revenue
monthly_revenue = df_cleaned.groupby('YearMonth')['Total Revenue'].sum()

# Plot
monthly_revenue.plot(kind='line', figsize=(12,6), marker='o', color='green')
plt.title('Monthly Revenue Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.tight_layout()
plt.savefig('monthly_revenue_trend.png' )
plt.show()
from google.colab import files
files.download('monthly_revenue_trend.png')
```

```python
top_customers = df_cleaned.groupby('CUSTOMERNAME')['Total Revenue'].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
top_customers.plot(kind='bar', figsize=(10,6), color='orange')
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Customer Name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('top_10_customers_bar.png')
plt.show()
from google.colab import files
files.download('top_10_customers_bar.png')
```

```python
# RFM ke liye latest date define karo
latest_date = df_cleaned['ORDERDATE'].max()

# Group by customer
rfm = df_cleaned.groupby('CUSTOMERNAME').agg({
    'ORDERDATE': lambda x: (latest_date - x.max()).days,  # Recency
    'ORDERNUMBER': 'nunique',                             # Frequency
    'Total Revenue': 'sum'                                # Monetary
})

# Rename columns
rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Display top few rows
rfm.head(10)
```

```python
# RFM Scoring (1 = worst, 5 = best)
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])  # Lower recency = better
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

# Combine into single RFM score
rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['RFM_Score'] = rfm[['R_Score', 'F_Score', 'M_Score']].astype(int).sum(axis=1)

rfm.head(10)
```

```python
def segment_customer(rfm):
    if rfm['RFM_Score'] >= 13:
        return 'Champions'
    elif rfm['RFM_Score'] >= 10:
        return 'Loyal Customers'
    elif rfm['RFM_Score'] >= 7:
        return 'Potential'
    elif rfm['RFM_Score'] >= 4:
        return 'At Risk'
    else:
        return 'Lost'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
rfm.head(10)
```

```python
# Count of customers in each segment
segment_counts = rfm['Segment'].value_counts()

# Plot pie chart
segment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(8,8))
plt.title('Customer Segments Distribution (RFM)')
plt.ylabel('')
plt.tight_layout()
plt.savefig('customer_segments_pie.png')
plt.show()
from google.colab import files
files.download('customer_segments_pie.png')
```

```python
# Save RFM table with segments to CSV
rfm.to_csv('RFM_Analysis_Output.csv')
```
